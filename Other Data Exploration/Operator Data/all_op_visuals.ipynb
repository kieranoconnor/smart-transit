```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read in the CSV file
alloperators = pd.read_csv("./alloperators.csv")
```

```python
plt.figure()
alloperators.groupby(by='op_type').kwh.mean().plot(kind='bar')
plt.title('Operator Type vs KWH Mean')
plt.xlabel('Operator Type')
plt.ylabel('KWH')
plt.show()
```

```python
# show the most common operator types
print("Operator types most often used")
alloperators['op_type'].value_counts().head(5)
```
