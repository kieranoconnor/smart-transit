```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read in the CSV file
df = pd.read_csv("./top10operators.csv")
display(df)
```

```python
#Plot the KWH vs Block for the top 15 operators
plt.figure()
df.groupby(by='block').kwh.mean().plot(kind='bar', xlabel='Block', ylabel='KWH')
plt.title('KWH vs Block')
```

```python
# Plot the Operator Type vs KWH STD
plt.figure()

df.groupby(by='op_type').kwh.std().plot(kind='bar')
plt.title('Operator Type vs KWH Standard Deviation')
plt.xlabel('Operator Type')

plt.ylabel('KWH')
plt.show()
```

```python
# Plot the Operator Type vs KWH Mean
plt.figure()
df.groupby(by='op_type').kwh.mean().plot(kind='bar')
plt.title('Operator Type vs KWH Mean')
plt.xlabel('Operator Type')
plt.ylabel('KWH')
plt.show()
```

```python
# Plot the Operator ID vs KWH STD
plt.figure()

df.groupby(by='op_id').kwh.std().plot(kind='bar')
plt.title('Operator ID vs KWH Standard Deviation')
plt.xlabel('Operator ID')

plt.ylabel('KWH')
plt.show()
```

```python
# Plot the Operator ID vs KWH Mean
plt.figure()

df.groupby(by='op_id').kwh.mean().plot(kind='bar')
plt.title('Operator ID vs KWH Mean')
plt.xlabel('Operator ID')

plt.ylabel('KWH')
plt.show()
```

```python
# Plot the Operator Type vs Average Temperature
plt.figure()

df.groupby(by='op_type').TAVG.mean().plot(kind='bar')
plt.title('Operator Type vs Temperature Mean')
plt.xlabel('Operator Type')

plt.ylabel('Avg Temperature')
plt.show()
```

```python
# Plot the Operator ID vs Average Temperature
plt.figure()

df.groupby(by='op_id').TAVG.mean().plot(kind='bar')
plt.title('Operator ID vs Temperature Mean')
plt.xlabel('Operator ID')

plt.ylabel('Avg Temperature')
plt.show()
```
