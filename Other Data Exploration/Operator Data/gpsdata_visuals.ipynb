```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./gpsdata.csv")
display(df)
```

```python
plt.figure()
df.groupby(by='BLOCK_NAME').PASSENGERS_IN.mean().plot(kind='bar', xlabel='Block', ylabel='Passengers')
plt.title('Block vs Passengers')
```

```python
plt.figure()
df.groupby(by='BLOCK_NAME').BLOCK_NAME.count().plot(kind='bar', xlabel='Block', ylabel='Count')
plt.title('Count vs Block')
```

```python
df_filtered = df[df['BLOCK_NAME'] >= 7000]
plt.figure()
df_filtered.groupby(by='BLOCK_NAME').BLOCK_NAME.count().plot(kind='bar', xlabel='Block', ylabel='Count')
plt.title('Count vs Block')
```

```python
plt.figure()
df_filtered.groupby(by='BLOCK_NAME').PASSENGERS_IN.mean().plot(kind='bar', xlabel='Block', ylabel='Passengers')
plt.title('Block vs Passengers')
```
